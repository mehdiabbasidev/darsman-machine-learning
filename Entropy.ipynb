# Entropy and Gain

```python
from math import log2

class0 = 5/100
class1 =95/100

entropy = -(class0 * log2(class0) + class1 * log2(class1))
print('Entropy:' , entropy)
```

```python
def entropy(class0, class1):
 return -(class0 * log2(class0) + class1 * log2(class1))
```

```python
class0 = 13 / 20
class1 = 7 / 20
s_entropy = entropy(class0, class1)
print(f'Dataset Entropy:{s_entropy:.3f}')
```

```python
s1_class0 = 7 / 8
s1_class1 = 1 / 8
s1_entropy = entropy(s1_class0, s1_class1)
print(f'Group1 Entropy:{s1_entropy:.3f}')
```

```python
s2_class0 = 10 / 12
s2_class1 = 2 / 12
s2_entropy = entropy(s2_class0, s2_class1)
print(f'Group2 Entropy:{s2_entropy:.3f}')
```

```python
gain = s_entropy - (8/20 * s1_entropy + 12/20 * s2_entropy)
print(f'Information Gain:{gain:.3f}')
```

# Entropy and Gini

```python
import matplotlib.pyplot as plt
import numpy as np

def gini(p):
   return (p)*(1 - (p)) + (1 - p)*(1 - (1-p))

def entropy(p):
   return - p*np.log2(p) - (1 - p)*np.log2((1 - p))


x = np.arange(0.0, 1.0, 0.01)
ent = [entropy(p) if p != 0 else None for p in x]
scaled_ent = [e*0.5 if e else None for e in ent]
gini=gini(x)

fig = plt.figure()
ax = plt.subplot(111)

for j, lab, ls, c, in zip(
      [ent, scaled_ent,gini ],
      ['Entropy', 'Entropy (scaled)', 'Gini'],
      ['-', '-', '--'],
      ['yellow', 'red', 'green']):
   line = ax.plot(x, j, label=lab, linestyle=ls, lw=1, color=c)

ax.legend(loc='upper left', bbox_to_anchor=(0.01, 0.85), ncol=1, fancybox=True, shadow=False)

ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')

plt.ylim([0, 1.1])
plt.xlabel('p(j=1)')
plt.ylabel('Impurity Index')
plt.show()
```
