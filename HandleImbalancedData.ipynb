# Oversampling

## Handle by SMOTE

```python
import pandas as pd
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns

X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, weights=[0.9], random_state=42)

df = pd.DataFrame(X, columns=['feature_1', 'feature_2'])
df['target'] = y
df.shape
print(df['target'].value_counts())

print(100*"*")

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
df_res = pd.DataFrame(X_res, columns=['feature_1', 'feature_2'])
df_res['target'] = y_res
df_res.shape

print(df_res['target'].value_counts())

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x='feature_1', y='feature_2', hue='target', data=df)
plt.title('Data Distribution Before SMOTE')
plt.subplot(1, 2, 2)
sns.scatterplot(x='feature_1', y='feature_2', hue='target', data=df_res)
plt.title('Data Distribution After SMOTE')
plt.show()
```

## ADASYN

```python
import pandas as pd
from sklearn.datasets import make_classification
from imblearn.over_sampling import ADASYN
import matplotlib.pyplot as plt
import seaborn as sns

X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, weights=[0.9], random_state=42)

df = pd.DataFrame(X, columns=['feature_1', 'feature_2'])
df['target'] = y
print(df['target'].value_counts())

print(100*"*")

adasyn = ADASYN(random_state=42)
X_res, y_res = adasyn.fit_resample(X, y)
df_res = pd.DataFrame(X_res, columns=['feature_1', 'feature_2'])
df_res['target'] = y_res
print(df_res['target'].value_counts())

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x='feature_1', y='feature_2', hue='target', data=df)
plt.title('Data Distribution Before ADASYN')
plt.subplot(1, 2, 2)
sns.scatterplot(x='feature_1', y='feature_2', hue='target', data=df_res)
plt.title('Data Distribution After ADASYN')
plt.show()
```

# Undersampling

## Random Under Sampling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, weights=[0.9], random_state=42)

df = pd.DataFrame(X, columns=['feature_1', 'feature_2'])
df['target'] = y
print(df['target'].value_counts())
df.shape
print(100*"*")

rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)
df_res = pd.DataFrame(X_res, columns=['feature_1', 'feature_2'])
df_res['target'] = y_res
print(df_res['target'].value_counts())

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x='feature_1', y='feature_2', hue='target', data=df)
plt.title('Data Distribution Before RandomUnder')
plt.subplot(1, 2, 2)
sns.scatterplot(x='feature_1', y='feature_2', hue='target', data=df_res)
plt.title('Data Distribution After RandomUnder')
plt.show()
```
