```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
tsne = TSNE(n_components=2, perplexity=5, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X)
X_tsne.shape
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_tsne[y == 0, 0], X_tsne[y == 0, 1], alpha=0.5, color='blue', marker='^', label=iris.target_names[0])
plt.scatter(X_tsne[y == 1, 0], X_tsne[y == 1, 1], alpha=0.5, color='red', marker='s', label=iris.target_names[1])
plt.scatter(X_tsne[y == 2, 0], X_tsne[y == 2, 1], alpha=0.5, color='green', marker='o', label=iris.target_names[2])
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.title('t-SNE on Iris Dataset')
plt.show()
```
